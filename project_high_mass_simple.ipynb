```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt('2024_12_21_c/history.data', names=True, skip_header=5)
data1 = np.genfromtxt('2024_12_21_f/LOGS_1221_f/history.data', names=True, skip_header=5)
```

```python
plt.plot(data['log_Teff'], data['log_L'], color = 'red', linewidth =0.95, linestyle='--', label=r'$5\ M_{\odot}$')
plt.plot(data1['log_Teff'], data1['log_L'], linewidth =0.95, label = r'$8\ M_{\odot}$')
plt.xlabel(r'$\log_{10} T_{\rm eff}~{[\rm K]}$')
plt.ylabel(r'$\log_{10} L~{[\rm L_\odot}]$')
plt.gca().invert_xaxis()
plt.legend()
#plt.savefig('d:/documents/Classes/Stellar/mesa_figure/high_mass_simple/HR_5_M_8_M.pdf')
plt.show()
```

```python
plt.plot(data['star_age']/1e7, data['log_Teff'],color = 'red', linewidth =0.95, linestyle='--', label=r'$5\ M_{\odot}$')
plt.plot(data1['star_age']/1e7, data1['log_Teff'], linewidth = 0.95, label = r'$8\ M_{\odot}$')
plt.xlabel(r'star age ($10^{7}$ yr)')
plt.ylabel(r'$\log_{10} T_{\rm eff}~{[\rm K]}$')
plt.legend()
#plt.savefig('d:/documents/Classes/Stellar/mesa_figure/high_mass_simple/star_age_T_eff_5_M_8_M.pdf')
plt.show()
```

```python
plt.plot(data['star_age']/1e7, data['log_R'],color = 'red', linewidth =0.95, linestyle='--', label=r'$5\ M_{\odot}$')
plt.plot(data1['star_age']/1e7, data1['log_R'], linewidth = 0.95, label = r'$8\ M_{\odot}$')
plt.xlabel(r'star age ($10^{7}$ yr)')
plt.ylabel(r'$\log_{10} R~{[\rm R_{\odot}]}$')
plt.legend()
#plt.savefig('d:/documents/Classes/Stellar/mesa_figure/high_mass_simple/star_age_R_5_M_8_M.pdf')
plt.show()
```
